### Importing the required package and library

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
```

### Function to Calculate the %

```python
def calculate_percentage(data):
    count = data.groupby("Category")["Category"].count()
    percentage = (count/len(data))*100
    return percentage
```

### Declare the setting for plotting e.g. color, font....

```python
#set font size of labels on matplotlib plots
plt.rc('font', size=16)

#set style of plots
sns.set_style('white')

customPalette = ['green', 'gold', 'blue', 'violet','lime','red']
sns.set_palette(customPalette)
sns.palplot(customPalette)
```

```python
### Function to plot the clustering result
```

```python
def plotClustering(data, continusAssessmentFinal, FinalResult, title):
    #plot data with seaborn (don't add a legend yet)
    facet = sns.lmplot(data=data, x=continusAssessmentFinal, y=FinalResult, hue='Category', 
                       fit_reg=False, legend=False)

    #add a legend
    leg = facet.ax.legend(bbox_to_anchor=[1, 0.75],
                             title="Legend", fancybox=True)
    #change colors of labels
    for i, text in enumerate(leg.get_texts()):
        plt.setp(text, color = customPalette[i])
    plt.title(title)
```

```python
def clusterViz(col, data):
    #cluster
    kmeans = KMeans(n_clusters = 3)
    lables = kmeans.fit_predict(data)  
    #color choice
    LABEL_COLOR_MAP = {
        0 : 'red',
        1 : 'green',
        2: 'yellow',
    }
    label_color = [LABEL_COLOR_MAP[l] for l in labels]
    
    #ploting the cluster
    plt.scatter(finalDS['Category'], data[col], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
    plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50, cmap='Spectral')

    Low = mpatches.Patch(color='red', label='Low')
    Medium = mpatches.Patch(color='Green', label='Medium')
    High = mpatches.Patch(color='yellow', label='High')
    
    y_values = ["Low SCORE", "Medium SCORE", "High SCORE"]
    y_axis = np.arange(1, 100, 35)

    plt.yticks(y_axis, y_values)
    
    plt.xlabel("Clusters")
    plt.legend(handles=[Low, Medium, High])

    plt.savefig('Cluster.pdf',dpi=1)
    plt.show()
```
